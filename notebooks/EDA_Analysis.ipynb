```python
# we have a file MachineLearningRating_v3.txt let load it and do an EDA on it
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# load the data
data = pd.read_csv('data/MachineLearningRating_v3.txt', sep='|')

# check the first 5 rows
data.sample(5)
```

```python
data.describe()
```

```python
#Calculate the variability for numerical features such as TotalPremium, TotalClaim

# Variability for TotalPremium
print('Variability for TotalPremium')
print('Mean:', data['TotalPremium'].mean())
print('Median:', data['TotalPremium'].median())
print('Standard Deviation:', data['TotalPremium'].std())
print('Variance:', data['TotalPremium'].var())
print('Range:', data['TotalPremium'].max() - data['TotalPremium'].min())
print('Interquartile Range:', data['TotalPremium'].quantile(0.75) - data['TotalPremium'].quantile(0.25))
print('Coefficient of Variation:', data['TotalPremium'].std() / data['TotalPremium'].mean())
print('Skewness:', data['TotalPremium'].skew())

print('---------------------------------')
# Variability for TotalClaims
print('Variability for TotalClaims')
print('Mean:', data['TotalClaims'].mean())
print('Median:', data['TotalClaims'].median())
print('Standard Deviation:', data['TotalClaims'].std())
print('Variance:', data['TotalClaims'].var())
print('Range:', data['TotalClaims'].max() - data['TotalClaims'].min())
print('Interquartile Range:', data['TotalClaims'].quantile(0.75) - data['TotalClaims'].quantile(0.25))
print('Coefficient of Variation:', data['TotalClaims'].std() / data['TotalClaims'].mean())
print('Skewness:', data['TotalClaims'].skew())
```

```python
#Review the dtype of each column 
data.dtypes
```

```python
# check for missing values
data.isnull().sum()
```

```python
for col in data.select_dtypes(include='object'):
    sns.countplot(data[col])
    plt.title(col)
    plt.show()
```

```python
# Explore relationships between the monthly changes TotalPremium and TotalClaims 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalPremium', y='TotalClaims', hue='PostalCode', data=data)
plt.xlabel('TotalPremium')
plt.ylabel('TotalClaims')
plt.title('TotalPremium vs TotalClaims by PostalCode')
plt.show()
```

```python
# Compare the change in insurance cover type, premium, auto make
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalPremium', y='TotalClaims', hue='CoverType', data=data)
plt.xlabel('TotalPremium')
plt.ylabel('TotalClaims')
plt.title('TotalPremium vs TotalClaims by CoverType')
plt.show()
```

```python
# Use box plots to detect outliers in numerical data
plt.figure(figsize=(10, 6))
sns.boxplot(data['TotalPremium'])
plt.title('Boxplot of TotalPremium')
plt.show()
```

```python
#Produce 3 creative and beautiful plots 
# Plot 1: Distribution of TotalPremium
plt.figure(figsize=(10, 6))
sns.histplot(data['TotalPremium'], bins=20, kde=True)
plt.title('Distribution of TotalPremium')
plt.show()

# Plot 2: Distribution of TotalClaims
plt.figure(figsize=(10, 6))
sns.histplot(data['TotalClaims'], bins=20, kde=True)
plt.title('Distribution of TotalClaims')
plt.show()

# Plot 3: Relationship between TotalPremium and TotalClaims using correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(data[['TotalPremium', 'TotalClaims']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between TotalPremium and TotalClaims')
plt.show()
```
